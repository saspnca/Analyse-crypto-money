```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#regression SKLearn
```

```python
np.random.seed(0)
m = 100 # creating 100 samples
X = np.linspace(0, 10, m).reshape(m, 1) #tableau en 2 dimensions
y = X + np.random.randn(m, 1) #modèle de regression linéaire
```

```python
plt.scatter(X, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X, y)
model.score(X, y) #coef de determination, le Rcarré de la technique des moindres carrés

predictions = model.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, c='r')
```

```python
#SVR regression si modèle de regression non linéaire
```

```python
np.random.seed(0)
m = 100 # creating 100 samples
X = np.linspace(0, 10, m).reshape(m, 1) #tableau en 2 dimensions
y = X**2 + np.random.randn(m, 1) #modèle de regression non linéaire
```

```python
from sklearn.svm import SVR
```

```python
model = SVR(C=100)
model.fit(X, y)
model.score(X, y) #coef de determination, le Rcarré de la technique des moindres carrés

predictions = model.predict(X)

plt.scatter(X, y)
plt.plot(X, predictions, c='r', lw=3)
```
