```python
import numpy as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
bitcoin = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
```

```python
bitcoin.head()
```

```python
bitcoin['Close'].plot(figsize=(9,6))
plt.show
```

```python
bitcoin.index
```

```python
bitcoin['2019-09']['Close'].plot()
```

```python
bitcoin.loc['2017':'2019','Close'].plot()
```

```python
# resample : regroupe mes données selon une fréquence (années, mois, jour, second etc...)
```

```python
bitcoin.loc['2019', 'Close'].resample('M').plot()
plt.show
```

```python
#évolution du bitcoin en 2019 en faisant la moyenne sur chaque mois !
```

```python
bitcoin.loc['2019', 'Close'].resample('M').mean().plot()
plt.show
```

```python
# compare le marché du bitcoin avec selon de l'ethereum
```

```python
ethereum = pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
```

```python
ethereum['2019']['Close'].plot()
```

```python
# j'assemble donc les 2 dataframe ensemble pour les comparés
```

```python
pd.merge(bitcoin, ethereum, on='Date', how='inner')
```

```python
btc_eth = pd.merge(bitcoin, ethereum, on='Date', how='inner', suffixes=('_btc', '_eth'))
```

```python
btc_eth.head()
```

```python
btc_eth[['Close_btc', 'Close_eth']].plot(subplots=True, figsize=(12,8))
```

```python
btc_eth[['Close_btc', 'Close_eth']].corr()
```
